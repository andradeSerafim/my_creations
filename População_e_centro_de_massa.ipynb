```python
def fazendo_matriz(ano) -> list:
  matriz = []
  for i in range(len(ano)):
    nova_linha = []
    for j in range(len(ano)):
      nova_linha.append(0)
    matriz.append(nova_linha)
  return matriz


fazendo_matriz([0, 0.1, 0.2, 0.3, 0.4])
```

```python
def diferenca_dividida(ano, populacao) -> list:
  matriz = fazendo_matriz(ano)
  # print('ano, populacao:', ano, populacao)  # Debug
  for i in range(len(populacao)):
    matriz[i][0] = populacao[i]
    # print(matriz)
    
  for i in range(len(populacao))[1:]:
    for j in range(len(ano))[1:]:
      numerador = matriz[i][j - 1] -  matriz[i - 1][j - 1]
      denominador = ano[i] - ano[i - j]
      # matriz[i][j] = numerador / denominador
      matriz[i][j] = round(numerador / denominador, 5)
      if i < j:
        matriz[i][j] = 0

  '''print()
  for i in range(len(populacao)):
    for j in range(len(ano)):
      print(matriz[i][j], end = '\t')
    print()
  print()'''

  return matriz

# diferenca_dividida([1970, 1971, 1972, 1973, 1974],
#                    [6000, 6200, 6600, 7200, 8000])
# diferenca_dividida([1970, 1971, 1972], [200.0, 400.0])
diferenca_dividida([1972, 1973, 1974], [6600, 7200, 8000])
# diferenca_dividida([1970, 1971, 1972, 1973, 1974],
#                    [6000, 6200, 6600, 7200, 8000])
# diferenca_dividida([-1, 0, 2],
#                    [4, 1,-1])
# diferenca_dividida([0, 8, 20, 30, 45],
#                    [0, 52.032, 160.45, 275.961, 370.276])
# diferenca_dividida([2, 2.2, 2.4, 2.6, 2.8],
#                    [0.5103757, 0.5207843, 0.5104147, 0.4813306, 0.435916])
# diferenca_dividida([0, 0.1, 0.2, 0.3, 0.4],
#                    [-0.3333, -0.292, -0.245, -0.1922, -0.1338])
```

```python
def polinomio(valor, anos, tabela) -> float:
  # print('tabela', tabela)  # Debug
  resultado = 0
  for i in range(len(tabela)+1):  # Linha da tabela
    monomio = 0
    for j in range(i+1):  # Coluna da tabela
      # print('i -1, j:', i -1, j)  # Debug
      if i -1 == j:
        # print('i -1, j:', i -1, j)  # Debug
        monomio += tabela[i-1][j]
        # print(f'tabela[{i-1}][{j}]', tabela[i-1][j])  # Debug
        for cada_ano in range(j):
          # print('j', j)
          # print(f'valor - anos[{cada_ano}]:', valor ,'-',anos[cada_ano])  # Debug
          monomio *= (valor - anos[cada_ano])
        # print()  # Debug
    resultado += monomio
  return resultado


# polinomio(2.5, [2, 2.2, 2.4, 2.6, 2.8],
#           diferenca_dividida([2, 2.2, 2.4, 2.6, 2.8],
#                              [0.5103757, 0.5207843, 0.5104147, 0.4813306, 0.435916]))
# polinomio(2.5, [2.2, 2.4, 2.6, 2.8],
#           diferenca_dividida([2.2, 2.4, 2.6, 2.8],
#                              [0.5207843, 0.5104147, 0.4813306, 0.435916]))
# polinomio(2.5, [2.2, 2.4, 2.6],
#           diferenca_dividida([2.2, 2.4, 2.6],
#                              [0.5207843, 0.5104147, 0.4813306]))
# polinomio(1, [0, 0.1, 0.2, 0.3, 0.4],
#           diferenca_dividida([0, 0.1, 0.2, 0.3, 0.4],
#                              [0.3333, -0.292, -0.245, -0.1922, -0.1338]))
polinomio(1973.5, [1972, 1973, 1974],
          diferenca_dividida([1972, 1973, 1974], [6600, 7200, 8000]))
# polinomio(1, [1972, 1973, 1974],
#           diferenca_dividida([1972, 1973, 1974],
#                              [6600, 7200, 8000]))
# polinomio(1, [1972, 1973],
#           diferenca_dividida([1972, 1973],
#                              [6600, 7200]))
```

```python
from typing import Tuple

Vector = Tuple[list]

def entradas_de_valores(ano: list, populacao: list) -> Vector:
  entrada: str = input('Digite valores de ano crescentes e de população ' \
                       'separados por espaço: ')
  
  if not ' ' in entrada:
    return ano, populacao
  
  novo_ano, nova_populacao = entrada.split()
  # print('novo_ano, nova_populacao', novo_ano, nova_populacao)  # Debug

  if not (novo_ano.replace('.','').isdecimal() and
      nova_populacao.replace('.','').isdecimal()):
    return ano, populacao

  ano.append(float(novo_ano))
  populacao.append(float(nova_populacao))
    
  ano, populacao = entradas_de_valores(ano, populacao)
  return ano, populacao


# entradas_de_valores([2], [0.5103757])
entradas_de_valores([1972], [6600])
```

```python
def centroide_triangulo_retangulo(base: float, altura: float, n_de_pontos: int):
  x = []
  f_x = []
  y = []
  f_y = []

  for i in range(n_de_pontos + 1):
    x += [i * base / n_de_pontos]

  for j in range(len(x)):  # Integral ¨y¨i dA = ¨y¨i * y dx
    f_x += [ (1/2) * (altura / base)**2 * (x[j])**2]
  
  for i in range(n_de_pontos + 1):
    y += [i * altura / n_de_pontos]

  for j in range(len(y)):  # Integral ¨x¨i dA = ¨x¨i * x dy
    f_y += [ (1/2) * (base / altura)**2 * (y[j])**2]

  
  return x, f_x, y, f_y


centroide_triangulo_retangulo(4, 2, 5)
# centroide_triangulo_retangulo(25, 20, 100)
# centroide_triangulo_retangulo(30, 30, 100)
```

```python
def area_corpo_monomio(base: float, altura: float, n_de_pontos: int, 
                          k: float = 1.0, g: float = 3.0):
  x = []
  f_x = []

  for i in range(n_de_pontos + 1):
    x += [ i * base / n_de_pontos ]
  
  for j in range(len(x)):
    f_x += [ (x[j] / k)**(1/g) ]

  return x, f_x


area_corpo_monomio(8, 2, 10, 1, 3)
```

```python
def centroide_corpo_monomio(base: float, altura: float, 
                               n_de_pontos: int, 
                               k: float = 1.0, g: float = 1.0):
    x = []
    f_x = []
    y = []
    f_y = []

    for i in range(n_de_pontos + 1):
      x += [i * base / n_de_pontos]

    for j in range(len(x)):  # Integral ¨y¨i dA = ¨y¨i * y dx
      f_x += [ (1/2) * (x[j] / k)**(2 / g)]
    
    for i in range(n_de_pontos + 1):
      y += [i * altura / n_de_pontos]

    for j in range(len(y)):  # Integral ¨x¨i dA = ¨x¨i * x dy
      f_y += [ (1/2) * (k**2) * (y[j])**(2*g) ]

    
    return x, f_x, y, f_y


# centroide_corpo_monomio(4, 2, 10, 0.00390625, 10)
# centroide_corpo_monomio(15, 10, 1000, 0.029928934724533184, 2.7)[3]
centroide_corpo_monomio(15, 10, 50, 0.029928934724533184, 2.7)[3]
```

```python
def trapezios(f_x: list, fim_de_x=1, inicio_de_x=0) -> float:
  if len(f_x) < 2:
    raise ValueError('Lista deve ter mais de um valor.')
  
  A = 0

  h = (fim_de_x - inicio_de_x) / (len(f_x) - 1)
  # print(h)  # Debug
  for i in range(len(f_x) - 1):
    A += h * (f_x[i] + f_x[i + 1]) / 2
  
  return A


# trapezios([1, 1.105, 1.221, 1.35, 1.492, 1.649, 1.822, 2.014, 2.226, 2.46,
#            2.718])
# trapezios([1])
# trapezios([2.718, 3.49, 4.482, 5.755, 7.389])
# trapezios([2.718, 3.211, 3.794, 4.482, 5.294, 6.255, 7.389])
# trapezios([1, 1.323, 1.581, 1.803, 2], 4, 1)
# trapezios([1, 1.225, 1.414, 1.581, 1.732, 1.871, 2], 4, 1)
# trapezios([0.707, 0.447, 0.354, 0.302, 0.267], 14, 2)
# trapezios([0.707, 0.5, 0.408, 0.354, 0.316, 0.289, 0.267], 14, 2)
# trapezios([1, 1/1.1, 1/1.2, 1/1.3, 1/1.4, 1/1.5, 1/1.6], 0.6)
# trapezios([1, 1.105, 1.221, 1.35, 1.492, 1.649, 1.822, 2.014, 2.226, 
#            2.46, 2.718])

# trapezios(centroide_triangulo_retangulo(4, 2, 40)[1], 4) / 4
# 4 - trapezios(centroide_triangulo_retangulo(4, 2, 40)[3], 2) / 4
# trapezios(centroide_triangulo_retangulo(25, 20, 100)[1], 25) / 250
# 25 - trapezios(centroide_triangulo_retangulo(25, 20, 100)[3], 20) / 250
# trapezios(centroide_triangulo_retangulo(30, 30, 100)[1], 30) / 450
# 30 - trapezios(centroide_triangulo_retangulo(30, 30, 100)[3], 30) / 450

# trapezios([0.0, 0.000032000000000000005, 0.00025600000000000004, 
#            0.0008639999999999999, 0.0020480000000000003,
#            0.004000000000000001, 0.006911999999999999,
#            0.010976000000000003, 0.016384000000000003,
#            0.023327999999999998, 0.03200000000000001, 0.042592,
#            0.05529599999999999, 0.070304, 0.08780800000000002,
#            0.10799999999999998, 0.13107200000000002, 0.15721600000000002,
#            0.18662399999999998, 0.21948800000000002, 0.25600000000000006,
#            0.29635199999999995, 0.340736, 0.389344, 0.44236799999999993,
#            0.5, 0.562432, 0.6298560000000001, 0.7024640000000002,
#            0.7804479999999998, 0.8639999999999999, 0.9533119999999999,
#            1.0485760000000002, 1.1499840000000001, 1.2577280000000002,
#            1.3719999999999997, 1.4929919999999999, 1.620896,
#            1.7559040000000001, 1.8982080000000001,2.0480000000000005,
#            2.2054719999999994, 2.3708159999999996, 2.544224, 2.725888,
#            2.9160000000000004, 3.114752, 3.3223359999999995, 3.5389439999999994,
#            3.7647679999999997, 4.0], 8)

# trapezios([0.0, 0.0037500000000000007,0.015000000000000003,0.03375,
#            0.06000000000000001,0.09375,0.135,0.18374999999999997,
#            0.24000000000000005,0.30375,0.375,0.4537500000000001,0.54,0.63375,
#            0.7349999999999999,0.84375,0.9600000000000002,1.0837499999999998,
#            1.215,1.35375,1.5,1.65375,1.8150000000000004,1.9837499999999997,
#            2.16,2.34375,2.535,2.7337500000000006,2.9399999999999995,3.15375,
#            3.375,3.6037500000000007,3.8400000000000007,4.083749999999999,
#            4.334999999999999,4.59375,4.86,5.133750000000001,5.415,
#            5.703749999999999,6.0], 8)

trapezios([0.0,  3.587324068671536e-63,  3.7615819226313243e-57,
           1.250822560397574e-53, 3.9443045261050636e-51,
           3.4211388289180104e-49, 1.3115825170914466e-47,
           2.8624071738327144e-46, 4.135903062765143e-45,
           4.361348592013147e-44,  3.587324068671532e-43], 4)

trapezios(centroide_corpo_monomio(15, 10, 1000, 0.029928934724533184, 
                                     2.7)[3], 10)
```

```python
def grafico_opcional(base, altura, n_de_pontos, k, g, centroide_x, centroide_y):
  plt.grid()
  lista_x = centroide_corpo_monomio(base, altura, n_de_pontos, k, g)[0]
  lista_f_x = [((i/k) ** (1 / g)) for i in lista_x]
  plt.plot(lista_x,
          lista_f_x)
  plt.fill(lista_x,
          lista_f_x,
          alpha=0.2)
  
  t = plt.Polygon([[0,0], [base, altura], [base,0]], alpha=0.2)
  plt.gca().add_patch(t)

  plt.axvline(0, color='k')
  plt.axhline(0, color='k')
  plt.plot(centroide_x, centroide_y, 'ro')
  plt.text(centroide_x + 0.1, centroide_y + 0.1, 
           s=f'({round(centroide_x,3)}, {round(centroide_y, 3)})', color='red')
  plt.title(f" x = {k} * y ^ {g}")
  plt.xlabel('x', size= 8)
  plt.ylabel('y', size= 8)
  plt.show()


grafico_opcional(16, 4, 1000, 0.25, 3, 7.238, 0.8)
```
